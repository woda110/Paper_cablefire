```python
import math
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
np.random.seed(0)


# def po(En,He,N):
#     for i in range (0,10000):
#         p = np.random.normal(En,He)
#         if p > 0:
#             a.append(p)
#         if len(a) == N:
#             return a

        
#定义正向(forward)云发生器
def forwardCloud(Ex,En,He,N):
    np.random.seed(0)
    a = np.random.normal(En,He,size=N)
    cloudpoints=[]

    
    for i in range(N):
        En2 = a[i] 
        x=np.random.normal(Ex,abs(En2))
        y=math.exp(-pow(x-Ex,2)/(2*pow(En2,2)))
        cloudpoints.append({'x':x,'y':y})
    return cloudpoints
```

```python
def sim(Ex,En,He,N):
    # 1.生成云滴
    res=forwardCloud(Ex,En,He,N)
    x=[]
    y=[]
    for r in res:
        x.append(r['x'])
        y.append(r['y'])
#     return x,y


    # 2.计算相似度
    
    X = np.array(x)
    u1=u2=u3= 0
    np.random.seed(0)
    Enn1 = np.random.normal(En_standard1,He_standard1,size=N)
    Enn2 = np.random.normal(En_standard2,He_standard2,size=N)
    Enn3 = np.random.normal(En_standard3,He_standard3,size=N)
    
    for i in range(len(X)):
        w1 = math.exp(-(X[i]-Ex_standard1)**2/(2*Enn1[i]**2))
        w2 = math.exp(-(X[i]-Ex_standard2)**2/(2*Enn2[i]**2))
        w3 = math.exp(-(X[i]-Ex_standard3)**2/(2*Enn3[i]**2))
        u1 += w1
        u2 += w2
        u3 += w3
        u = u1/N+u2/N+u3/N
    return [u1/N/u,u2/N/u,u3/N/u]

```

```python
# 1.电流标准云
Ex_standard1,En_standard1,He_standard1 = 5, 1.6667, 0.1
Ex_standard2,En_standard2,He_standard2 = 14, 1.3333, 0.1
Ex_standard3,En_standard3,He_standard3  = 22, 1.3333, 0.1
```

```python
sim(9.9,0.3,0.1,3000)
```
